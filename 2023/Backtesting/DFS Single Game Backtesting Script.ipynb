```python
def player_scores(pred_lineup, test, game_index):
    actual_lineups =pred_lineup
    actual_lineups = actual_lineups[['0','1','2','3','4']]
    actual_lineups0 = actual_lineups[['0']]
    actual_lineups0 = actual_lineups0.rename(columns = {'0':'Player'})
    actual_lineups1 = actual_lineups[['1']]
    actual_lineups1 = actual_lineups1.rename(columns = {'1':'Player'})
    actual_lineups2 = actual_lineups[['2']]
    actual_lineups2 = actual_lineups2.rename(columns = {'2':'Player'})
    actual_lineups3 = actual_lineups[['3']]
    actual_lineups3 = actual_lineups3.rename(columns = {'3':'Player'})
    actual_lineups4 = actual_lineups[['4']]
    actual_lineups4 = actual_lineups4.rename(columns = {'4':'Player'})


    actual_lineups_final = actual_lineups0.append(actual_lineups1)
    actual_lineups_final = actual_lineups_final.append(actual_lineups2)
    actual_lineups_final = actual_lineups_final.append(actual_lineups3)
    actual_lineups_final = actual_lineups_final.append(actual_lineups4)

    actual_lineups_final = actual_lineups_final.drop_duplicates()
    test = test[test['game_index_x'] == game_index]
    test = test[['Player ID + Player Name','Points']]
    
    player_scores = pd.merge(actual_lineups_final, test, left_on = 'Player',
             right_on = 'Player ID + Player Name', how = 'outer')
    return player_scores    
    
    
```

```python
def top_winning_lineups(temp100, game_index, winning_thres):
    
    actual_lineups =temp100[temp100['game_index']==game_index]
    actual_lineups = actual_lineups[['adj_total_score','mvp_player','player0','player1','player2','player3','player4']]
    actual_lineups = actual_lineups.sort_values('adj_total_score', ascending =False)
    actual_lineups_top10 = actual_lineups[actual_lineups['adj_total_score'] >= winning_thres]#.head(150)
    actual_lineups_top10 = actual_lineups_top10.reset_index()
    return actual_lineups_top10
```
